```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('population.csv')
```

```python
data.head(200)
```

```python
countries = [	'Afghanistan','China', 'India', 'Brazil', 'United States']
population_data = data[data['Country Name'].isin(countries)]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(population_data['Country Name'], population_data['2020']) # Or another year
plt.title('Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```
